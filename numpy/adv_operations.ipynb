## Advance operations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 120000, 170000, 200000],
    [2, 160000, 130000, 180000, 190000],
    [3, 170000, 140000, 190000, 180000],
    [4, 180000, 150000, 200000, 170000],
    [5, 190000, 160000, 210000, 160000]
])


print("=== Sales Data ===")
print("sales_data.shape: ", sales_data.shape)

print("SAmple data of first three rows: ", sales_data[0:3])

```

```python
# total sales per year

total_sales = np.sum(sales_data[:, 1:], axis=0)
print("total_sales: ", total_sales)

# min sales per product
total_sales_per_product = np.min(sales_data[:, 1:], axis=1)
print("total_sales_per_product: ", total_sales_per_product)

# max sales per year
total_sales_per_product = np.max(sales_data[:, 1:], axis=0)
print("total_sales_per_year: ", total_sales_per_product)

# average sales per product
average_sales_per_product = np.mean(sales_data[:, 1:], axis=1)
print("average_sales_per_product: ", average_sales_per_product)
```

```python
cumilative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("cumilative_sales: ", cumilative_sales)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumilative_sales, axis=0))
plt.title("Cumilative Sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
