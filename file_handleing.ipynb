```python
from typing import TextIO

data : TextIO = open("./input.txt") # conectivity with input file
print(data.read())

data.close() # close the file
```

```python
with open("./input.txt") as data:
    print(type(data))
    print(data.readline()) # read the first line
    
    
# with helps us to close the file automatically 
    
```

```python
with open("./input.txt") as data:
    print(type(data))
    print(data.readlines()) # read all/rest of the lines 
    

    
```

```python
with open("./input.txt", "r+") as data:
    print(data.read())
    data.write("Hello World")
    data.seek(0)
    
    print(data.read())
    
    
```

```python
with open("./inputt.txt", "w") as data:  
    data.write("Hello World")
    
# if file does not exist, it will create a new file
# dont append, it will overwrite the file     
```

```python
with open("./inputt.txt", "a") as data:  
    data.write("Hello World")
    
# if file does not exist, it will create a new file
# add the content at the end of the file    
```

```python
with open("./inputt.txt", "rb+") as data:  
    print(data.read())
```

```python
from typing import List
import PyPDF2

def read_pdf(pdf_path: str) -> List[str]:
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = [reader.pages[i].extract_text() for i in range(len(reader.pages))]
    return text

pages: List[str] = read_pdf("./data.pdf")
pages
```
